# PDF Report

```python
import fpdf
```

```python
pdf = fpdf.FPDF(orientation='portrait', unit='mm', format='A4')
date_from = "2020/01/01"
date_to = "2020/03/19"

# First page
pdf.add_page()

# Title image
pdf.image("../reporting/images/mercadona-title.jpg", x=pdf.w/4, y=30, w=106.3, h=70.9)
pdf.ln(100)

# Title text
pdf.set_font('helvetica', "B", size=18)
pdf.cell(w=0, txt="Mercadona Prices Report", fill=False, align="C")
pdf.ln(10)
pdf.set_font('helvetica', size=16)
pdf.cell(w=0, txt=f"{date_from} - {date_to}", fill=False, align="C")

# Second Page


pdf.output("../reporting/reports/Mercadona_prices_report.pdf")
```
